```python
import torch
from transformers import AutoImageProcessor, SiglipForImageClassification
from PIL import Image

# The working Indian/Western specialist
MODEL_NAME = "prithivMLmods/Indian-Western-Food-34"

print(f"Loading Specialist Model: {MODEL_NAME}")
food_processor = AutoImageProcessor.from_pretrained(MODEL_NAME)
# Note: This model uses Siglip architecture
food_model = SiglipForImageClassification.from_pretrained(MODEL_NAME)
food_model.eval()

print("‚úÖ Specialist model loaded.")

def get_specialist_food_name(pil_image):
    inputs = food_processor(images=pil_image, return_tensors="pt")
    with torch.no_grad():
        outputs = food_model(**inputs)
    predicted_class_idx = outputs.logits.argmax(-1).item()
    return food_model.config.id2label[predicted_class_idx]
```

```python
from transformers import AutoProcessor, AutoModelForVision2Seq

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
DTYPE = torch.float16 if torch.cuda.is_available() else torch.float32

GENERALIST_MODEL_ID = "Salesforce/blip-image-captioning-large"

print(f"Loading Generalist Model: {GENERALIST_MODEL_ID}")
blip_processor = AutoProcessor.from_pretrained(GENERALIST_MODEL_ID)
blip_model = AutoModelForVision2Seq.from_pretrained(GENERALIST_MODEL_ID).to(DEVICE, dtype=DTYPE)
blip_model.eval()

print("‚úÖ Generalist model loaded.")

def get_generalist_description(pil_image):
    prompt = "<DETAILED_CAPTION>"
    inputs = blip_processor(images=pil_image, text=prompt, return_tensors="pt").to(DEVICE)
    for k, v in inputs.items():
        if v.is_floating_point(): inputs[k] = v.to(dtype=DTYPE)

    with torch.no_grad():
        # Added repetition_penalty to stop "> > >" errors
        out = blip_model.generate(**inputs, max_new_tokens=128, repetition_penalty=1.5, no_repeat_ngram_size=2)

    text = blip_processor.decode(out[0], skip_special_tokens=True)
    return text.replace("< detailed _ caption >", "").replace("< caption >", "").strip()
```

```python
from langgraph.graph import StateGraph, END
from typing import TypedDict, Dict, Any
import traceback

class GraphState(TypedDict):
    image: Image.Image
    specialist_guess: str
    generalist_description: str
    final_json_report: Any
    final_summary: str

def run_specialist(state: GraphState) -> Dict[str, str]:
    print("--- üßë‚Äç‚öïÔ∏è SPECIALIST NODE ---")
    try:
        guess = get_specialist_food_name(state["image"])
        print(f"Guess: {guess}")
        return {"specialist_guess": guess}
    except Exception:
        return {"specialist_guess": "Error"}

def run_generalist(state: GraphState) -> Dict[str, str]:
    print("--- üïµÔ∏è GENERALIST NODE ---")
    try:
        desc = get_generalist_description(state["image"])
        print(f"Desc: {desc}")
        return {"generalist_description": desc}
    except Exception:
        return {"generalist_description": "Error"}
```

```python
def make_fusion_prompt(specialist_guess, generalist_description):
    schema = """
Return a single JSON object with:
- dish_name: string (Specific cultural name)
- dish_confidence: number 0-1
- ingredients: array of {name, confidence}
- portion_size: {size, grams_estimate}
- nutrition_estimate: {calories, protein_g, carbs_g, fat_g}
- health_flags: list of strings
- diet_advice: list of strings
    """

    return f"""
You are an expert AI Nutritionist.

EVIDENCE:
1. Specialist Model Guess: "{specialist_guess}"
2. Generalist Model Description: "{generalist_description}"

---
**PHASE 1: GATEKEEPER (Is it food?)**
Analyze the "Generalist Model Description".

**PASSPHRASE RULE (IMPORTANT):**
If the description contains words like **"food", "plate", "dish", "meal", "curry", "rice", "bread"**, or names of ingredients, **IMMEDIATELY ACCEPT IT as food.**

**REJECTION RULES (Only apply if Passphrase Rule fails):**
1. If the text is purely nonsense symbols (e.g. "> > >", "|||"), REJECT.
2. If it clearly describes a NON-FOOD object (e.g. "man's face", "screenshot", "car", "keyboard"), REJECT.

IF REJECTED: Output ONLY plain text: "This does not appear to be a food item."

**PHASE 2: IDENTIFICATION (Who to trust?)**
You must resolve the conflict between the models using this **HIERARCHY OF TRUST**:
1. **The Specialist Model** is the authority on NAMES (e.g. "Masala Dosa", "Biriyani").
2. **The Generalist Model** is the authority on VISUAL COMPONENTS (e.g. "Fish", "Rice", "Soup").

**CRITICAL RULES:**
- If Specialist says "Pakoda" but Generalist clearly says "Fish", synthesize them: **"Fish Pakoda"**.
- If Specialist says "Masala Dosa" and Generalist says "Crepe" or "Food", use **"Masala Dosa"**.

**PHASE 3: GENERATION**
STEP 1: Generate JSON enclosed in [JSON] and [/JSON] tags.
STEP 2: Write a summary after the JSON.

SCHEMA:
{schema}

OUTPUT:
[JSON]
{{...}}
[/JSON]
[Summary]
"""
```

```python
import os
import re
import json
import google.generativeai as genai

# ‚ö†Ô∏è PASTE YOUR KEY HERE
os.environ["GEMINI_API_KEY"] = "Get yours from Google AI Studio"
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

def extract_json(s):
    m = re.search(r"\[JSON\]\s*(\{[\s\S]*?\})\s*\[/JSON\]", s)
    if not m: m = re.search(r"(\{[\s\S]*\})", s)
    if not m: return None, s
    try:
        return json.loads(m.group(1)), s[m.end(0):].strip()
    except:
        return None, s

def call_fusion_llm(prompt):
    try:
        # The correct working model with high token limit
        model = genai.GenerativeModel('gemini-2.5-flash')
        config = genai.types.GenerationConfig(temperature=0.0, max_output_tokens=8192)
        resp = model.generate_content(prompt, generation_config=config)

        # Safe access to text
        if resp.candidates and resp.candidates[0].content.parts:
            return resp.candidates[0].content.parts[0].text
        return '{"error": "Empty response"}'
    except Exception as e:
        return f'{{"error": "{str(e)}"}}'

def run_gemini_brain(state: GraphState) -> Dict[str, str]:
    print("--- üß† BRAIN NODE ---")
    prompt = make_fusion_prompt(state["specialist_guess"], state["generalist_description"])
    raw_out = call_fusion_llm(prompt)
    json_rpt, summary = extract_json(raw_out)
    return {"final_json_report": json_rpt, "final_summary": summary}
```

```python
workflow = StateGraph(GraphState)
workflow.add_node("specialist", run_specialist)
workflow.add_node("generalist", run_generalist)
workflow.add_node("gemini_brain", run_gemini_brain)

workflow.set_entry_point("specialist")
workflow.add_edge("specialist", "generalist")
workflow.add_edge("generalist", "gemini_brain")
workflow.add_edge("gemini_brain", END)

app = workflow.compile()
print("‚úÖ App Compiled.")
```

```python
import nest_asyncio
import gradio as gr
import json
import time

# 1. Patch the event loop to prevent "RuntimeError"
nest_asyncio.apply()

# 2. Force close any stuck servers
gr.close_all()

# --- HELPER: Beautiful Markdown Formatter ---
def format_to_markdown(json_data, summary):
    if not json_data or "error" in json_data:
        return f"### ‚ö†Ô∏è Could not analyze image.\n\n**Reason:** {json_data.get('error', 'Unknown error') if json_data else 'No data'}"

    dish = json_data.get("dish_name", "Unknown Dish")

    nutri = json_data.get("nutrition_estimate", {})
    cals = nutri.get("calories", "N/A")
    prot = nutri.get("protein_g", "N/A")
    carbs = nutri.get("carbs_g", "N/A")
    fat = nutri.get("fat_g", "N/A")

    portion = json_data.get("portion_size", {})
    size_desc = portion.get("size", "N/A")
    grams = portion.get("grams_estimate", "N/A")

    md = f"""
# üçΩÔ∏è **{dish}**
### {summary}

---

## üìä Nutritional Profile
| Nutrient | Amount |
| :--- | :--- |
| **Calories** | **{cals} kcal** |
| ü•© Protein | {prot}g |
| üçû Carbs | {carbs}g |
| ü•ë Fat | {fat}g |

* **Portion Estimate:** {size_desc} (~{grams}g)

---

## ü•ó Ingredients Identified
"""
    for ing in json_data.get("ingredients", []):
        md += f"* {ing.get('name')} _({int(ing.get('confidence', 0)*100)}% confidence)_\n"

    md += "\n--- \n\n## üö¶ Health Analysis\n"

    flags = json_data.get("health_flags", [])
    if flags:
        for flag in flags:
            md += f"* ‚ö†Ô∏è **{flag}**\n"
    else:
        md += "* ‚úÖ No specific health warnings.\n"

    md += "\n## üí° Dietitian's Advice\n"
    for tip in json_data.get("diet_advice", []):
        md += f"1. {tip}\n"

    md += f"\n\n_Sources used: {', '.join(json_data.get('sources', []))}_"
    return md

# --- MAIN LOGIC WITH PROGRESS BAR ---
# We add 'progress=gr.Progress()' to the inputs
def analyze_food(image, progress=gr.Progress()):
    if image is None:
        return "Please upload an image to start."

    # 1. Start Visual Feedback
    progress(0.1, desc="üöÄ Starting Image Analysis...")
    print("--- üöÄ FRONTEND REQUEST STARTED ---")

    try:
        # 2. Invoke the LangGraph Pipeline
        # Since LangGraph runs as a block, we wrap it in a progress update
        progress(0.3, desc="üîç Specialist & Generalist Models Scanning...")

        # This is the heavy lifting step
        final_state = app.invoke({"image": image})

        progress(0.7, desc="üß† Gemini Nutritionist Thinking...")

        # 3. Extract Results
        rpt = final_state.get("final_json_report")
        summary = final_state.get("final_summary", "")

        progress(0.9, desc="‚ú® Formatting Report...")

        # 4. Check for Rejection/Error
        if not rpt or "error" in rpt:
            error_msg = rpt.get('error') if rpt else "Unknown Error"
            display_text = summary if summary else f"‚ùå Error: {error_msg}"
            return f"### {display_text}"

        # 5. Success! Return beautiful Markdown
        return format_to_markdown(rpt, summary)

    except Exception as e:
        return f"### ‚ùå Critical System Error\n\n{str(e)}"

# --- UI LAYOUT ---
# Using the 'Soft' theme for a cleaner look
with gr.Blocks(theme=gr.themes.Soft(), title="AI Nutritionist") as demo:
    gr.Markdown("# ü•ó AI Nutritionist")
    gr.Markdown("Upload a photo of your meal (Indian or Western) for a detailed breakdown.")

    with gr.Row():
        # Left: Input
        with gr.Column(scale=1):
            img_in = gr.Image(type="pil", label="Upload Food Photo", height=400)
            # The button
            btn = gr.Button("Analyze Food üöÄ", variant="primary", size="lg")

        # Right: Output
        with gr.Column(scale=1):
            report_out = gr.Markdown(label="Nutrition Report", value="Upload an image to see results here...")

    # Connect button to function
    btn.click(analyze_food, inputs=[img_in], outputs=[report_out])

# Launch
print("Starting server... Watch for the public link below!")
demo.launch(share=True, debug=True)
```
